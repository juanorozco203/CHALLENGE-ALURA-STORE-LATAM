### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda1.head()
```

#1. Análisis de facturación



```python
ingreso_tienda1 = tienda1['Precio'].sum()
ingreso_tienda2 = tienda2['Precio'].sum()
ingreso_tienda3 = tienda3['Precio'].sum()
ingreso_tienda4 = tienda4['Precio'].sum()

print(f"Ingresos Tienda 1: ${ingreso_tienda1}")
print(f"Ingresos Tienda 2: ${ingreso_tienda2}")
print(f"Ingresos Tienda 3: ${ingreso_tienda3}")
print(f"Ingresos Tienda 4: ${ingreso_tienda4}")
```

# 2. Ventas por categoría

```python


print("Tienda 1")
print(tienda1['Categoría del Producto'].value_counts())

print("\nTienda 2")
print(tienda2['Categoría del Producto'].value_counts())

print("\nTienda 3")
print(tienda3['Categoría del Producto'].value_counts())

print("\nTienda 4")
print(tienda4['Categoría del Producto'].value_counts())
```

# 3. Calificación promedio de la tienda


```python


print("Calificación promedio Tienda 1:", tienda1['Calificación'].mean())
print("Calificación promedio Tienda 2:", tienda2['Calificación'].mean())
print("Calificación promedio Tienda 3:", tienda3['Calificación'].mean())
print("Calificación promedio Tienda 4:", tienda4['Calificación'].mean())
```

# 4. Productos más y menos vendidos

```python


print("=== Tienda 1 ===")
print("Más vendido:", tienda1['Producto'].value_counts().idxmax())
print("Menos vendido:", tienda1['Producto'].value_counts().idxmin())

print("\n=== Tienda 2 ===")
print("Más vendido:", tienda2['Producto'].value_counts().idxmax())
print("Menos vendido:", tienda2['Producto'].value_counts().idxmin())

print("\n=== Tienda 3 ===")
print("Más vendido:", tienda3['Producto'].value_counts().idxmax())
print("Menos vendido:", tienda3['Producto'].value_counts().idxmin())

print("\n=== Tienda 4 ===")
print("Más vendido:", tienda4['Producto'].value_counts().idxmax())
print("Menos vendido:", tienda4['Producto'].value_counts().idxmin())
```

# 5. Envío promedio por tienda

```python
print("Envío promedio Tienda 1:", tienda1['Costo de envío'].mean())
print("Envío promedio Tienda 2:", tienda2['Costo de envío'].mean())
print("Envío promedio Tienda 3:", tienda3['Costo de envío'].mean())
print("Envío promedio Tienda 4:", tienda4['Costo de envío'].mean())
```

# ANALISIS GRAFICO


```python
import matplotlib.pyplot as plt
```

#
 GRÁFICO 1: Ingresos totales por tienda

```python
ingresos = [tienda1['Precio'].sum(), tienda2['Precio'].sum(),
            tienda3['Precio'].sum(), tienda4['Precio'].sum()]
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

plt.bar(tiendas, ingresos, color=['blue', 'green', 'orange', 'red'])
plt.title('Ingresos Totales por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Ingresos')
plt.show()
```

#
GRÁFICO 2:  Calificación promedio

```python
calificaciones = [tienda1['Calificación'].mean(), tienda2['Calificación'].mean(),
                  tienda3['Calificación'].mean(), tienda4['Calificación'].mean()]

plt.barh(tiendas, calificaciones, color=['blue', 'green', 'orange', 'red'])
plt.title('Calificación Promedio por Tienda')
plt.xlabel('Calificación')
plt.show()
```

#
GRÁFICO 3:  Participación en ingresos

```python
plt.pie(ingresos, labels=tiendas, autopct='%1.1f%%')
plt.title('Participación en Ingresos Totales')
plt.show()
```
